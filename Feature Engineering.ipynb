# **Detecting missing values numerically**

- The 1st step is to detect count/percentage of missing values in every column of dataset.
- This will give an idea about the distribution of missing values

```python
missing_count=df.isnull().sum()
missing_percent=df.isnull().mean()*100
missing_data=pd.DataFrame({'missing count':missing_count,'missing percent':missing_percent})
missing_data=missing_data[missing_data['missing count']>0]
missing_data
```

# **Detecting missing data visually using missingno library**

- To graphically analyse the missingness of data.let's use a library called missingno.
- It is a package for graphical analysis of missing values.
- To use this library `missingno`,we need import.

```python
import missingno as msno
msno.bar(df)
```

```python
msno.matrix(df)
```

```python
msno.matrix(df.sample(100))
```
