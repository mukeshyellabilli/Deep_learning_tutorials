# **loss and cost function**

**loss functions types**
- Mean Absolute Error=1/n summation abs(y-y^)
- Mean Square Error= 1/n summation (y-y^)^2
- Log loss(or)Binary Cross Entropy= 1/n summation ylog(y^)+(1-y).log(1-y^)

Note:
- Individual errors are called loss
- Cummulative error is called cost function

```python
import numpy as np
```

**Mean Absolute Error**

```python
y_pred=np.array([1,1,0,0,1])
y_true=np.array([0.30,0.7,1,0,0.5])
```

```python
def mae(y_true,y_pred):
    total_error=0
    for yt,yp in zip(y_true,y_pred):
        total_error+=abs(yt-yp)
    print("Total error:",total_error)
    mae=total_error/len(y_true)
    print("Mean absolute error:",mae)
    return mae
```

```python
mae(y_true,y_pred)
```

```python
np.mean(np.abs(y_pred-y_true))#mean absolute error
```

```python
np.sum(np.abs(y_pred-y_true))# total error
```

```python
def log_loss(y_true,y_pred):
    epslion = 1e-15
    y_pred_new = [max(i,epslion) for i in y_pred]
    y_pred_new = [min(i,1-epslion) for i in y_pred_new]
    y_pred_new = np.array(y_pred_new)
    return -np.mean(y_true*np.log(y_pred_new)+(1-y_true)*np.log(1-y_pred_new))
```

```python
log_loss(y_true,y_pred)
```

**Mean Square Error**

```python
def mse(y_true,y_pred):
    total_error=0
    for yt,yp in zip(y_true,y_pred):
        total_error+=(yt-yp)**2
    print("Total error:",total_error)
    mse=total_error/len(y_true)
    print("Mean square error:",mse)
    return mse
```

**without using numpy**

```python
mse(y_true,y_pred)
```

**with using numpy**

```python
np.mean(np.square(y_pred-y_true))#mean square error
```

```python
np.sum(np.square(y_pred-y_true))# total error
```
