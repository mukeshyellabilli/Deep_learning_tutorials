### **What is BERT?**

BERT (Bidirectional Encoder Representations from Transformers) is a state-of-the-art natural language processing (NLP) model developed by **Google AI**. It is based on the Transformer architecture and was introduced in 2018. BERT is designed to understand the context of words in a sentence by looking at both **left** and **right** contexts simultaneously, which makes it bidirectional.

---

### **Key Features of BERT**
1. **Bidirectional Context Understanding**: Unlike traditional models like Word2Vec or GloVe that look at text sequences either left-to-right or right-to-left, BERT considers the **entire context** of a word, improving its understanding of sentence meaning.

2. **Pre-training and Fine-tuning**:
   - **Pre-training**: BERT is trained on a large corpus of text (like Wikipedia) to learn general language representations.
   - **Fine-tuning**: These pre-trained representations can be fine-tuned on specific tasks like question answering, sentiment analysis, or named entity recognition.

3. **Transformer-based Architecture**: BERT uses the Transformer architecture, which relies on the **self-attention mechanism** to weigh the importance of different words in a sequence, enabling it to capture complex relationships.

---

### **How BERT Works**
BERT is pre-trained on two key tasks:
1. **Masked Language Model (MLM)**:
   - A percentage of the input tokens are masked, and the model predicts the missing words.
   - Example: For the input "The cat is [MASK]", BERT predicts "black" or other likely words.

2. **Next Sentence Prediction (NSP)**:
   - The model is trained to understand relationships between two sentences by predicting if one sentence logically follows the other.

---

### **Applications of BERT**
1. **Text Classification**: Sentiment analysis, spam detection.
2. **Question Answering**: Extracting precise answers from a passage of text.
3. **Named Entity Recognition (NER)**: Identifying entities like names, dates, and locations in text.
4. **Text Summarization**: Generating concise summaries of long documents.
5. **Language Translation**: Improving machine translation systems.

---

### **Advantages of BERT**
- Improved accuracy in understanding the **context** and meaning of ambiguous or complex text.
- Versatility in addressing a wide range of NLP tasks through fine-tuning.
- Open-sourced, allowing developers to adapt and use it in various applications.

### **Sentence Example**
Imagine we want to predict the meaning of the word **"bank"** in two different sentences:

1. **"He sat by the river bank."**
2. **"She went to the bank to deposit money."**

---

### **How BERT Understands Context**
- Traditional models might assign the same meaning to "bank" in both sentences, as they often don't consider the surrounding words.
- **BERT**, however, looks at the **entire sentence** to understand the context:
  - In sentence 1, BERT sees words like "river" and "sat," understanding that "bank" refers to the edge of a river.
  - In sentence 2, BERT notices words like "deposit" and "money," identifying that "bank" here refers to a financial institution.

---

### **Masked Word Example**
Using the **Masked Language Model (MLM)**:

If you provide the sentence:
**"The dog is [MASK] in the garden."**

BERT predicts likely words to replace "[MASK]":
- "playing"
- "sitting"
- "running"

Here is the information on basic uncased BERT model,

https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4

It uses L=12 hidden layers (i.e., Transformer blocks), a hidden size of H=768, and A=12 attention heads. This model has been pre-trained for English on the Wikipedia and BooksCorpus.

```python
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

```python
bert_preprocess_model = hub.KerasLayer(preprocess_url)
```

```python
text_test = ['nice movie indeed','I love python programming']
text_preprocessed = bert_preprocess_model(text_test)
text_preprocessed.keys()
```

```python
text_preprocessed['input_mask']
```

```python
text_preprocessed['input_type_ids']
```

```python
text_preprocessed['input_word_ids']
```

**101 --> CLS token**

```python
bert_model = hub.KerasLayer(encoder_url)
```

```python
bert_results = bert_model(text_preprocessed)
```
