# **sigmoid function**

```python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(90)
```

```python
sigmoid(40)
```

```python
sigmoid(6)
```

```python
sigmoid(10)
```

```python
sigmoid(-63)
```

# **tanh function**

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-10)
```

```python
tanh(0)
```

```python
tanh(100)
```

# **relu function**

```python
def relu(x):
    return max(0,x)
```

```python
relu(-100)
```

```python
relu(70)
```

```python
relu(1)
```

# **leaky_relu**

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(69)
```

```python
leaky_relu(-56)
```

```python
leaky_relu(0)
```
